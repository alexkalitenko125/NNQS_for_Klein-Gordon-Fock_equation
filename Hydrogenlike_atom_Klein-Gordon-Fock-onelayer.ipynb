# Relativistic spinless particle in a Coulomb potential

In relativistic quantum mechanics, for a free particle with rest mass energy, the relationship between energy $E$ and momentum $p$
\begin{equation}
E^2 = p^2 c^2 + m^2c^4,
\end{equation}
where  $c$ is the speed of light in a vacuum, $m$ is the mass of the particle. With a potential $V(x)$,  we obtain
\begin{equation}
[E-V]^2 = p^2 c^2 + m^2c^4.
\end{equation}

The corresponding wave equation in one dimension is
\begin{equation}
[E-V]^2 \psi(x) = \left[ -\hbar^2 c^2 \frac{d^2}{dx^2}+ m^2c^4 \right]\psi(x), 
\end{equation}
where $\hbar$ is the reduced Planck constant, $\psi(x)$ is the wavefunction. 

It is a so-called relativistic Klein--Gordon--Fock equation. We can write it in the Schrödinger-like equation 

\begin{equation}
E_{eff} \psi(x) = \left[ -\frac{\hbar^2}{2m}  \frac{d^2}{dx^2}+ V_{eff}(x) \right]\psi(x), 
\end{equation}
where we denote 
\begin{equation}
E_{eff} = \frac{E^2-m^2c^4}{2mc^2},
\end{equation}

\begin{equation}
V_{eff} = \frac{2EV(x) - V^2(x)}{2mc^2},
\end{equation}

\begin{equation}
\hat{H}_{eff} = \left[ -\frac{\hbar^2}{2m}  \frac{d^2}{dx^2}+ V_{eff}(x) \right],
\end{equation}


\begin{equation}
E_{eff} = \frac{\langle \psi|\hat{H}_{eff}|\psi\rangle}{\langle \psi|\psi \rangle}\geq E_0,
\end{equation}
$E_0$ is the minimum energy for our system.

The hydrogenlike atom can be solved analytically with the Schrödinger equation. We will write down the form of some functions $P_{nl}(r)$ ("reduced radial functions" $P_{nl}(r)= rR_{nl}$) for some values $n, l$:

$$ P_{10}(r) = 2(\alpha Z)^{3/2}re^{-\alpha Zr}, $$

$$ P_{20}(r) = \frac{1}{\sqrt{2}}(\alpha Z)^{3/2}r\bigg(1-\frac{1}{2}\alpha Zr\bigg)e^{-\alpha Zr/2}, $$

$$ P_{21}(r) = \frac{1}{2\sqrt{6}}(\alpha Z)^{5/2}r^2e^{-\alpha Zr/2}. $$

Below is a simple Python implementation of these functions, which we will use later for comparative analysis, and the energies for the KGF equation.

```python
P_10 = lambda r : 2.*np.power(Z*alpha, 1.5)*r*np.exp(-Z*r*alpha)
P_20 = lambda r : (1/np.sqrt(2.))*np.power(Z*alpha, 1.5)*r*(1-0.5*Z*alpha*r)*np.exp(-Z*alpha*r/2.)
P_21 = lambda r : (1/(2*np.sqrt(6.)))*np.power(Z*alpha, 2.5)*np.power(r, 2)*np.exp(-alpha*Z*r/2.)
d1 = 0.5 - np.sqrt(0.25-Z**2*alpha**2) 
d2 = 1.5 - np.sqrt(1.5**2-Z**2*alpha**2)
E0_th = 1/np.sqrt(1+Z**2*alpha**2/(1-d1)**2)-1
E1_th = 1/np.sqrt(1+Z**2*alpha**2/(2-d2)**2)-1
E2_th = 1/np.sqrt(1+Z**2*alpha**2/(2-d1)**2)-1
#E0_th = -2.664e-5*Z**2
#E1_th = -6.66e-6*Z**2

r_grid = np.linspace(train_a, train_b, 1000)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.plot(r_grid, P_10(r_grid), color='r', label='$P_{10}$')
ax.plot(r_grid, P_20(r_grid), color='b', linestyle='-.', label='$P_{20}$')
ax.plot(r_grid, P_21(r_grid), color='g', linestyle='-.', label='$P_{21}$')
ax.legend()
fig.canvas.draw()
```

## The neural network

```python
class NNQS(nn.Module):
    def __init__(self, W1, B1, W2):
        super().__init__()
        
        # Set the layers 
        self.lc1 = nn.Linear(in_features=Nin, 
                             out_features=Nhid, 
                             bias=True)  
        self.actfun = nn.Tanh()        # activation function
        self.lc2 = nn.Linear(in_features=Nhid, 
                             out_features=Nout, 
                             bias=True) 

        # Set the parameters 
        with torch.no_grad():
            self.lc1.weight = nn.Parameter(W1)
            self.lc1.bias = nn.Parameter(B1)
            self.lc2.weight = nn.Parameter(W2)

    # Set the output
    def forward(self, x): 
        out = self.lc2(self.actfun(self.lc1(x)))
        return out
    
net = NNQS(W1, B1, W2).to(device)
```

We use AdamW as our optimizer.

```python
optimizer = torch.optim.AdamW(params=net.parameters(), weight_decay=0.1, lr=lr)
```

## The loss function

The loss function is calculated as follows

$$ m = 1, c = 1, \hbar = 1, e^2 = \alpha  $$

$$\newcommand{\bra}[1]{{ \left\langle #1 \right|} }
\newcommand{\ket}[1]{{ \left|  #1 \right\rangle } }
\mathcal{L}=\frac{\bra{\psi_{nl,\theta}}\hat{H}_{eff}\ket{\psi_{nl,\theta}}}{\bra{\psi_{nl,\theta}}\psi_{nl,\theta}\rangle} + L_{2,regul}. $$

$$ \hat{H}_{eff} = -\frac{1}{2}\frac{d^2}{dx^2} +\frac{1}{2r^2}l(l+1)+ V_{eff}$$

$$ E_{eff} = \frac{E^2-1}{2} $$
$$ V_{eff} = \frac{2EV(x)-V^2(x)}{2} $$


Since we are working in spherical coordinates and given that the interaction is Coulombian, the above integrals

$$ \bra{\psi_{nl,\theta}}\hat{H}_{eff}\ket{\psi_{nl,\theta}} = \int_0^\infty \bigg\{\frac{1}{2}\bigg(\frac{d\psi_{nl,\theta}(r)}{dr}\bigg)^2+\psi_{nl,\theta}^2(r)(\frac{1}{2r^2}l(l+1)+ V_{eff})\bigg\}dr $$

With discretization
$$ \bra{\psi_{nl,\theta}}\hat{H}_{eff}\ket{\psi_{nl,\theta}} \approx \sum_{i=1}^{N_r} w_i \bigg[\frac{1}{2}\bigg\{\frac{d\psi_{nl,\theta}(r)}{dr}\bigg\}^2_{r=r_i}+\psi_{nl,\theta}^2(r_i)(\frac{1}{2r_i^2}l(l+1)+ V_{eff})\bigg], $$

$$\bra{\psi_{nl,\theta}}\psi_{nl,\theta}\rangle\approx \sum_{i=1}^{N_r} w_i \psi_{nl,\theta}^2(r_i)\equiv N.$$

At this stage, we take an additional step and add the envelope to the NNQS when calculating losses. The envelope is a function that multiplies NNQS, and its purpose is to impose boundary conditions. In our case, we should require that $\psi_{nl}(0)=0$, and we can achieve this in exactly the same way as it would be done analytically:

$$\psi_{nl}(r) \longrightarrow  \psi_{nl}(r)r^{l+1}. $$
We use $(\frac{1}{2r^2}l(l+1)+ V_{eff})$ as $U_{eff}$. 

```python
def loss_fn(l, E): 
    lambd = 0
    X_det = X.clone().detach()
    psi = net(X)*X.pow(l+1)
    psi2 = (net(X_det)*X_det.pow(l+1)).pow(2)
    dpsi_dr, = grad(outputs=psi, 
                    inputs=X,
                    grad_outputs=torch.ones_like(psi), 
                    create_graph=True)   
    N = torch.tensordot(psi2, wi) 
    U_eff = torch.tensordot(psi2 * ( - E*alpha*Z/X_det + l*(l+1)/(2*X_det.pow(2)) - (alpha*Z)**2/(2*X_det.pow(2)) ), wi) / N  
    K = (1/2) * torch.tensordot(dpsi_dr.pow(2), wi) / N
    regul = torch.tensordot(net.lc1.weight,net.lc1.weight) 
    + torch.tensordot(net.lc2.weight,net.lc2.weight)  # regularization
    
    return K + U_eff + lambd*regul, U_eff, K, psi, psi / torch.sqrt(N)
```

Now let's move on to the first excited state with $n=2, l=0$. The approach is almost identical to the procedure of finding the ground state wavefunction, but now we want to find an independent orthogonal solution

$$ \langle\psi_{10,\theta}|\psi_{20,\theta'}\rangle = 0, $$

where $\psi_{10,\theta}$ and $\psi_{20,\theta'}$ are the wavefunctions of the ground and first excited states respectively. The obvious way to impose orthogonality is to add this condition to the loss function using a Lagrange multiplier:

$$ \mathcal{L} = \frac{\langle{\psi_{20,\theta'}}|\hat{H}_{eff}|{\psi_{20,\theta'}}\rangle}{\langle\psi_{20,\theta'}|\psi_{20,\theta'}\rangle} + \lambda\frac{\langle\psi_{\theta,10}|\psi_{20,\theta'}\rangle^2}{\langle\psi_{20,\theta'}|\psi_{20,\theta'}\rangle\langle\psi_{10,\theta}|\psi_{10,\theta}\rangle} + L_{2,regul}, $$

where $\lambda$ is the Lagrange multiplier. 

```python
def loss_fn_excited(l, E):  
    lambd = 0
    X_det = X.clone().detach()
    psi = net1(X)*X.pow(l+1)
    psi2 = (net1(X_det)*X_det.pow(l+1)).pow(2)
    dpsi_dr, = grad(outputs=psi, 
                    inputs=X,
                    grad_outputs=torch.ones_like(psi), 
                    create_graph=True)   
    N = torch.tensordot(psi2, wi) 
    U_eff = torch.tensordot(psi2 * ( - E*alpha*Z/X_det + l*(l+1)/(2*X_det.pow(2)) - (alpha*Z)**2/(2*X_det.pow(2)) ), wi) / N  
    K = (1/2) * torch.tensordot(dpsi_dr.pow(2), wi) / N
    

    num = (torch.tensordot(psi*psi_10, wi)).pow(2)
    den = (torch.tensordot(psi2, wi)) * (torch.tensordot(psi_10.pow(2), wi))
    O = lambda_ * (num / den)
    regul = torch.tensordot(net1.lc1.weight,net1.lc1.weight) 
    + torch.tensordot(net1.lc2.weight,net1.lc2.weight) #regularization
    
    

    return K + U_eff + O +lambd*regul, U_eff, K, psi, psi / torch.sqrt(N)
```
